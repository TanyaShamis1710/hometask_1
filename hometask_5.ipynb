```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import random as rng

def Brownian_motion(n):
  
    steps_x=[]
    steps_y=[]
    pace_x=[]
    pace_y=[]
    i=0
    steps_x.append(rng(1))
    steps_y.append(rng(1))
    pace_x.append(0)
    pace_y.append(0)

    if (steps_x[i]< 0.5):
      pace_x.append(-1)
      #print("kciskd")
    else:
      pace_x.append(1)
    
    if (steps_y[i] < 0.5):
      pace_y.append(-1)
    else:
      pace_y.append(1)
    while((np.cumsum(pace_x)[i+1]!=0) or (np.cumsum(pace_y)[i+1]!=0)):
      #print("hhfhfhfhfhfhfhf")
      i+=1
      steps_x.append(rng(1))
      steps_y.append(rng(1))

      
      if (steps_x[i]< 0.5):
        pace_x.append(-1)
      else:
        pace_x.append(1)
    
      
      if (steps_y[i] < 0.5):
        pace_y.append(-1)
      else:
        pace_y.append(1)
    

      if ((np.cumsum(pace_x)[i]==n) or (np.cumsum(pace_y)[i]==n)):
        steps_x=[]
        steps_y=[]
        pace_x=[]
        pace_y=[]
        i=0
        steps_x.append(rng(1))
        steps_y.append(rng(1))
        pace_x.append(0)
        pace_y.append(0)

        if (steps_x[i]< 0.5):
          pace_x.append(-1)
          #print("kciskd")
        else:
          pace_x.append(1)
        
        if (steps_y[i] < 0.5):
          pace_y.append(-1)
        else:
          pace_y.append(1)
        
      
    plt.plot(np.cumsum(pace_x), np.cumsum(pace_y))
    plt.axis('equal')
    plt.scatter(np.cumsum(pace_x)[i+1],np.cumsum(pace_y)[i+1])
    plt.show()

a = Brownian_motion(100)
```

```
n = 1000
u1, u2 = np.random.rand(2, n)
cov=np.array([[2,1],[1,4]])
mean=np.array([3,6])
ro=cov[0,1]/(np.sqrt(cov[0,0]*cov[1,1]))
theta = 2 * np.pi * u2
eix = np.exp(1j*theta)
r = np.sqrt((-2 *(1-ro**2)/(1-2*ro*eix.real*eix.imag))* np.log(u1))

x = (r * eix.real+mean[0])*np.sqrt(cov[0,0])
y = (r * eix.imag+mean[1])*np.sqrt(cov[1,1])

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].hist2d(x, y, bins=30)
ax[0].axis('equal')
ax[1].hist(x, bins=30)
ax[2].hist(y, bins=30)
ax[0].set_title("2d histogram")
ax[1].set_title("Histogram of x")
ax[2].set_title("Histogram of y")
plt.show()

```

```
import random

cache = {}

def count_partitions(n, limit):
    if n == 0:
        return 1
    if (n, limit) in cache:
        return cache[n, limit]
    x = cache[n, limit] = sum(count_partitions(n-k, k) for k in range(1, min(limit, n) + 1))
    return x

def random_partition(n):
    a = []
    limit = n
    total = count_partitions(n, limit)
    which = random.randrange(total)
    while n:
        for k in range(1, min(limit, n) + 1):
            count = count_partitions(n-k, k)
            if which < count:
                break
            which -= count
        a.append(k)
        limit = k
        n -= k
    return a

random_partition(100)
```
